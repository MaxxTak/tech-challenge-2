```python
import os
import pymysql.constants
from dotenv import load_dotenv, find_dotenv

# Automatically find and load the .env file from the parent/root directory
load_dotenv(find_dotenv())

# Get database connection settings
db_host = os.getenv("DB_HOST", "localhost")
db_port = int(os.getenv("DB_PORT", 3306))
db_user = os.getenv("DB_USER", "root")
db_password = os.getenv("DB_PASSWORD", "password")
db_database = os.getenv("DB_DATABASE", "tech")

# Resolve SQL script path relative to this notebook
sql_file_path = os.path.join("..", "database", "br_inep_avaliacao_alfabetizacao_uf.sql")

print(f"Connecting to database '{db_database}' on '{db_host}:{db_port}' as '{db_user}'...")

# Connect to the database enabling multiple statement execution
connection = pymysql.connect(
    host=db_host,
    port=db_port,
    user=db_user,
    password=db_password,
    database=db_database,
    client_flag=pymysql.constants.CLIENT.MULTI_STATEMENTS
)

try:
    with open(sql_file_path, "r", encoding="utf-8") as f:
        sql_script = f.read()
        
    with connection.cursor() as cursor:
        print("Executing SQL script to populate database...")
        cursor.execute(sql_script)
    connection.commit()
    print("Database table created and populated successfully!")
except Exception as e:
    print(f"Error occurred: {e}")
finally:
    connection.close()
```
